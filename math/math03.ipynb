```python
#расчет длины вектора

import math

def calculate_len(vector):
    return math.sqrt(sum([v**2 for v in vector]))
v1=[10,10,10]
v2=[0,0,-10]

print(calculate_len(v1))
print(calculate_len(v2))


```

```python
#рисуем окружность
#(x-a)**2+(y-b)**2=R**2
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math

a=1
b=1
R=4
i=np.linspace(-4,6,100000)
x=[]
y1=[]
y2=[]
for x_ in i:
    try:
        y1_=math.sqrt(R**2-(x_-a)**2)+b
        y2_=-math.sqrt(R**2-(x_-a)**2)+b
        x.append(x_)
        y1.append(y1_)
        y2.append(y2_)
    except:
        pass
        
plt.grid(True)
plt.plot(a,b,color="red", marker="o")
plt.plot(x,y1, color="green")
plt.plot(x,y2, color="green")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.axis("scaled")
plt.show()

```

```python
#рисуем эллипс
#(x-x0)**2/a**2+(y-y0)**2/b**2=1
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
import random as r

x0,y0=1,2
a,b=5,3
c=math.sqrt(a**2-b**2)
i=np.linspace(-6,8,100000)
x, y1, y2 = [],[],[]
for x_ in i:
    try:
        y1_=b*math.sqrt(1-(x_-x0)**2/a**2)+y0
        y2_=-b*math.sqrt(1-(x_-x0)**2/a**2)+y0
        x.append(x_)
        y1.append(y1_)
        y2.append(y2_)
    except:
        pass

plt.grid(True)

ir = r.randint(0,len(x))
xr,yr = x[ir], y1[ir]
plt.plot([xr, -c+x0], [yr, y0], color='blue')
plt.plot([xr, c+x0], [yr, y0], color='blue')
    
plt.plot(x,y1, color="green")
plt.plot(x,y2, color="green")
plt.plot(x0,y0,color="red", marker="o")
plt.plot(-c+x0,y0,color="blue", marker=".")
plt.plot(c+x0,y0,color="blue", marker=".")
plt.plot(xr,yr,color="blue", marker=".")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.axis("scaled")
plt.show()
```

```python
# рисуем гиперболу 
# (тут решил делать без смещений, так как в задании не просят, а делается аналогично окружности и эллипсу)
# x**2/a**2-y**2/b**2=1

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math

a,b=1,2
x1=np.linspace(a,5,100)
y1, y2 = [],[]
for x_ in x1:
    y1_=b*math.sqrt(x_**2/a**2-1)
    y2_=-b*math.sqrt(x_**2/a**2-1)
    y1.append(y1_)
    y2.append(y2_)
x2=[-x_ for x_ in x1]    

plt.grid(True)
plt.plot(x1,y1, color="green")
plt.plot(x1,y2, color="green")
plt.plot(x2,y1, color="green")
plt.plot(x2,y2, color="green")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

```python
#рисуем 2 параллельные плоскости
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax=Axes3D(fig)
A,B,C,D1,D2=1,1,1,0,4
X=np.arange(-5,5,1)
Y=np.arange(-5,5,1)
X, Y =np.meshgrid(X, Y)
Z1=A/C*X+B/C*Y+D1/C
Z2=A/C*X+B/C*Y+D2/C
ax.plot_wireframe(X,Y,Z1, color="green")
ax.plot_wireframe(X,Y,Z2, color="orange")
show()
```

```python
# рисуем плоскости второго порядка
# пусть это будет эллипсоид окруженный двухполостным гиперболоидом
import matplotlib.pyplot as plt
import numpy as np
import math
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax=Axes3D(fig)
X=np.arange(-2,2, .05)
Y=np.arange(-2,2, .05)
X, Y =np.meshgrid(X, Y)
a,b,c = 2,2,2
"""
Уверен что есть способ избавиться от предупреждения интерпретатора 
о вычислении корня из отрицательных чисел, но сходу загуглить не получилось
"""
Z1_ellipsoid=np.sqrt((1-X**2/a**2-Y**2/b**2))*c
Z2_ellipsoid=-np.sqrt((1-X**2/a**2-Y**2/b**2))*c
Z1_hyperboloid=np.sqrt((1+X**2/a**2+Y**2/b**2))*c
Z2_hyperboloid=-np.sqrt((1+X**2/a**2+Y**2/b**2))*c

ax.plot_wireframe(X,Y,Z1_hyperboloid, color="blue")
ax.plot_wireframe(X,Y,Z2_hyperboloid, color="blue")
ax.plot_wireframe(X,Y,Z1_ellipsoid, color="green")
ax.plot_wireframe(X,Y,Z2_ellipsoid, color="green")
show()
```

```python
# рисуем y(x) = k∙cos(x – a) + b
x=np.arange(-3*np.pi,3*np.pi, .1)
plt.grid(True)
plt.xlim(-2*np.pi,2*np.pi)
plt.ylim(-3,3)
#k=1,a=0,b=0
plt.plot(x, np.cos(x), color="blue")
#k=2,a=0,b=-1
plt.plot(x, 2*np.cos(x)-1, color="green")
#k=1.5,a=pi,b=1
plt.plot(x, 1.5*np.cos(x+np.pi)+1, color="red")
plt.show()
```

```python
#переводим полярные координаты в декартовы
def from_polar_to_dekart(r, a):
    return (r*cos(a), r*sin(a))

#пример для пифагорова треугольника
print(from_polar_to_dekart(5,np.arctan(3/4)))
```

```python
#решаем систему уравнений
# y = x2 – 1
# exp(x) + x∙(1 – y) = 1
x=np.linspace(-4, 6, 100)
plt.plot(x, x**2-1, 'green')
plt.plot(x, (np.exp(x)+x-1)/x, 'blue')
plt.show()

from scipy.optimize import fsolve
def equations(p):
    x,y=p
    return (x**2-y-1, exp(x)+x*(1-y)-1)

x1, y1 = fsolve(equations, (-2,0))
print(x1, y1)
x2, y2 = fsolve(equations, (2,5))
print(x2, y2)
x3, y3 = fsolve(equations, (5,20))
print(x3, y3)
```
